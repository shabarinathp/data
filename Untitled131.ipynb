#insert at end

```python
class node:
    def __init__(self, dataval=None):
        self.dataval=dataval
        self.nextval=None

class linklist:
    def __init__(self):
        self.headval=None

    def linklist_print(self):
        printval=self.headval
        while printval is not None:
            print(printval.dataval)
            printval=printval.nextval
    def insert_beg(self,newdata):
        Newnode=node(newdata)
        Newnode.nextval=self.headval
        self.headval=Newnode
    def insert_end(self,newdata):
        Newnode=node(newdata)
        if self.headval is None:
            self.headval=Newnode
            return
        l=self.headval
        while(l.nextval):
            l=l.nextval
        l.nextval=Newnode
```

```python
x1=linklist()
x1.headval=node("Jan")
x2=node("feb")
x3=node("mar")
x1.headval.nextval=x2
x2.nextval=x3
```

```python
x1.linklist_print()
```

```python
x1.insert_end("Apr")
```

```python
x1.linklist_print()
```
